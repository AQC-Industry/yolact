## Correcting annotations files

```python
path = "../qualitex_full_size/annotations/train.json"
with open(path, "r") as j:
    data = json.load(j)
df_anno = pd.DataFrame(data["annotations"])
df_info = pd.DataFrame(data["info"])
df_licence = pd.DataFrame(data["licences"])
df_images = pd.DataFrame(data["images"])
df_cat = pd.DataFrame(data["categories"])
df_anno
```

```python
df_anno_clean = df_anno.copy()
for i in range(len(df_anno)):
    if len(df_anno["segmentation"].values[i]) == 0:
        df_anno_clean.drop(i, inplace=True)
    elif len(df_anno["segmentation"].values[i][0]) < 5:
        df_anno_clean.drop(i, inplace=True)
```

```python
file = {}
result = df_info.to_json(orient="records")
parsed = json.loads(result)
file["info"] = parsed

result = df_licence.to_json(orient="records")
parsed = json.loads(result)
file["licences"] = parsed

result = df_info.to_json(orient="records")
parsed = json.loads(result)
file["info"] = parsed

result = df_images.to_json(orient="records")
parsed = json.loads(result)
file["images"] = parsed

result = df_cat.to_json(orient="records")
parsed = json.loads(result)
file["categories"] = parsed

result = df_anno_clean.to_json(orient="records")
parsed = json.loads(result)
file["annotations"] = parsed

# Save JSON file
output_file = path
with open(output_file, 'w') as json_file:
    json.dump(file, json_file)
```
